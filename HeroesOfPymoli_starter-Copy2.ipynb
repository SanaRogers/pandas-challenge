```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pd_df = pd.read_csv(file_to_load)
pd_df.head()
```

## Player Count

* Display the total number of players


```python
# creating variable for number of players to print out as a dataframe
total_players = len(pd.unique(pd_df["SN"]))
tp_df = pd.DataFrame({
    "Total Players":[total_players]})
tp_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# set variables of price and revenue by calculating the average and sums respectively
average_price = pd_df["Price"].mean()
total_revenue = pd_df["Price"].sum()

# set variables of purchaes and unique items by counting unique values of each 
total_purchases = len(pd.unique(pd_df["Purchase ID"]))
unique_items = len(pd.unique(pd_df["Item ID"]))

# created dataframe from values rendered above
pa_df = pd.DataFrame({"Number of Unique Items":[unique_items], 
                      "Average Price":[average_price],
                      "Number of Purchases":[total_purchases],
                      "Total Revenue": total_revenue})

# converted price and revenue to a currency value 
pa_df["Average Price"] = pa_df["Average Price"].map("${:,.2f}".format)
pa_df["Total Revenue"] = pa_df["Total Revenue"].map("${:,.2f}".format)

pa_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# created dataframe to show total counts and % of players by gender and put in ascending order of total count
gender_df = pd_df[["SN", "Gender", "Age"]].drop_duplicates()
gender_group = gender_df["Gender"].value_counts()

# calculated percent of gender
perc_gender = gender_group/total_players

# dataframe for gender breakdown by total count and % of players within each group 
renamed_gender = pd.DataFrame({"Total Count":gender_group, "Percentage of Players": perc_gender})

# changed number to percentage 
renamed_gender["Percentage of Players"] = renamed_gender["Percentage of Players"].map("{:,.2%}".format)
renamed_gender
```

* Create a summary data frame to hold the results

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# grouped by screename to get the necessary values 
purchase_count = pd_df.groupby("SN").count()["Item ID"]
average_purch_price = pd_df.groupby("SN").mean()["Price"]

# created variables and added sorting 
total_purchase = purchase_count*average_purch_price
top_spenders = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":average_purch_price,"Total Purchase Value":total_purchase})
top_spenders_final = top_spenders.sort_values("Total Purchase Value",ascending=False)

# changed numerical vaues to currency
top_spenders_final["Average Purchase Price"] = top_spenders_final["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_final["Total Purchase Value"] = top_spenders_final["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_final.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# grouped items by Item ID and Item Name by finding item's relative price and value by multiplying 
# by purchase count
purchase_count = pd_df.groupby(["Item ID","Item Name"]).count()["Price"]
total_value = pd_df.groupby(["Item ID", "Item Name"]).sum()["Price"]
price = total_value/purchase_count

# created dataframe to be ascending by purchase count
most_popular = pd.DataFrame({"Purchase Count":purchase_count,
                             "Item Price":price,
                             "Total Purchase Value":total_value})
most_popular_final = most_popular.sort_values("Purchase Count",ascending=False)

# changing number format to currency 
most_popular_final["Item Price"] = most_popular_final["Item Price"].map("${:,.2f}".format)
most_popular_final["Total Purchase Value"] = most_popular_final["Total Purchase Value"].map("${:,.2f}".format)
most_popular_final["Purchase Count"] = most_popular_final["Purchase Count"].astype(str).str.replace(".0","")
most_popular_final.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# grouped by item ID and Item Name 
purchase_count = pd_df.groupby(["Item ID","Item Name"]).count()["Price"]

# created variables and put in a dataframe to ascending by Total Purchase Value
total_value = pd_df.groupby(["Item ID", "Item Name"]).sum()["Price"]
price = total_value/purchase_count

most_popular = pd.DataFrame({"Purchase Count":purchase_count,
                             "Item Price":price,
                             "Total Purchase Value":total_value})
most_popular_final = most_popular.sort_values("Total Purchase Value",ascending=False)

# changed number formatting to currency and removed tenths from Purchase Count column 
most_popular_final["Item Price"] = most_popular_final["Item Price"].map("${:,.2f}".format)
most_popular_final["Total Purchase Value"] = most_popular_final["Total Purchase Value"].map("${:,.2f}".format)
most_popular_final["Purchase Count"] = most_popular_final["Purchase Count"].astype(str).str.replace(".0","")
most_popular_final.head()
```
